  1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.
     - Covariance is a measure of how two variables change together, while correlationis a standardized measure of the strength and direction of the linear relationship between two variables, covariance has no standardized scale its value can range from negative infinty to positive infinity, and its magnitude is difficult to interpret on its own, covariance is unitless and is standardized to a scale between -1 and +1, making its magnitude easily interpretable, interpretation is a positive covariance indicates that the variables trend to move in the same direction, while a negative value indicates they trend to move in opposite direction, but the strength of this relationship cannot be determined from the value alone, correlation not only shows the direction but also shows the strength, values close to +1 and -1 indicates a strong linear relationship, while values near 0 indicate a weak or no linear relationship.

  4. Explain the difference between Pearson's correlation coefficient and Spearman's rank correlation coefficient. When would you prefer to use Spearman's correlation?
     - Pearson's correlation measures linear relationships between continuous, normally distributed data, while spearman's correlation measures monotonic (consistently increasing or decreasing, but not necessarily linear) relationships between ranked or ordinal data, pearson's product-moment correlation coefficient evaluates the strength and direction of a linear relationship between two continuous variables that are typically assumed to be jointly normally distributed, it uses the raw data values in its calculation, spearman's rank correlation coefficient is a non-parametric measure that evaluates the monotonic relationship between two continuous or ordinal variables, it is based on the ranked values of the data rather than the raw data itself, one would prefer to use Spearman's correlation in the following situations: a> when the data is ordinal or ranked (e.g., Likert scales), b> when the data is continuous but not normally distributed, c> when the relationship between variables is monotonic but not necessarily linear, d> when the data contains significant outliers that could distort Pearson's analysis, as Spearman's is more robust to extreme values.
